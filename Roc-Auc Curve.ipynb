```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples = 1000, n_classes=2, n_features=20,random_state=1)

print(X.shape,  y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(X_train,y_train)

pred_prob1 = model1.predict_proba(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier(n_neighbors=4)
model2.fit(X_train,y_train)

pred_prob2 = model2.predict_proba(X_test)
```

```python
print(pred_prob1)
```

```python
print(pred_prob2)
```

```python
from sklearn.metrics import roc_curve

fp1r, tp1r, _ = roc_curve(y_test,pred_prob1[:,1],pos_label=1)
fp2r, tp2r, _ = roc_curve(y_test,pred_prob2[:,1],pos_label=1)
```

```python
from sklearn.metrics import roc_auc_score

auc_score1 = roc_auc_score(y_test,pred_prob1[:,1])
auc_score2 = roc_auc_score(y_test,pred_prob2[:,1])

print(auc_score1,auc_score2 )
```

```python
import matplotlib.pyplot as plt


plt.plot(fp1r, tp1r,linestyle="--",color="orange",label="Logistic Regression")
plt.plot(fp2r, tp2r,linestyle="--",color="green",label="KNN")
```
